```python
import numpy as np
import matplotlib.pyplot as plt
```

Task : Predict firm's gross profit


- I start by performing 1000 simulations of company's expected revenues

First I assume company's last year's revenue(170M$),std deviation(20M$) and defne the no. of iterations of observervations I want to produce(1000).

```python
rev_m = 170
std_m = 20
n_iter = 1000
```

We predict the revenue first with a numpy normal distribution

```python
rev = np.random.normal(rev_m,std_m,n_iter)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
#plt.hist(rev,bins=100)
plt.show()
```

We see the data between the defined std deviation

I assume and take COGS(cost of goods) is ~60% of the company's revenue

```python
cogs = - (rev * np.random.normal(0.6,0.1))
plt.figure(figsize=(15,6))
plt.plot(cogs)
plt.show()
```

```python
cogs.mean(),cogs.std()
```

Gross Profit = revenue - COGS,since we take cogs as negative, I perform '+' in the below equation

I also want the the max,min,mean,std deviation of the goss profit

```python
gross_profit = rev + cogs
plt.figure(figsize=(15,6))
plt.plot(gross_profit)
plt.show()
```

```python
max(gross_profit),min(gross_profit),gross_profit.mean(),gross_profit.std()
```

```python
plt.figure(figsize =(15,6))
plt.hist(gross_profit,bins=20)
plt.show()
```
